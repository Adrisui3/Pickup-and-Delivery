```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

files = os.listdir('../weights_output')
print("Datasets loaded: ", files)
```

```python
for file in files:
    df = pd.read_csv('../weights_output/' + file)
    sns.lineplot(data = df).set_title(file)
    plt.show()
```
